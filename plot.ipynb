```python
import matplotlib.pyplot as plt
from google.colab import files
from pylab import rcParams
csfont = {'fontname':'Comic Sans MS'}
hfont = {'fontname':'Helvetica'}
rcParams['figure.figsize'] = 12, 5   
scheme = ['Patel et al. [9]','Dua et al. [17]','Baghestani et al.[18]','Garg et al. [19]','Sadhukhan et al. [26]','Qiu et al. [20]','Ours']
comm_cost = [800,2100,1696,864,1600,1280,992]
comp_cost = [8.9438,17.8448,8.9464,13.548,4.5036,8.9339,4.5003]
sec_level = [8,9,8,10,7,11,13]
fig, ((ax1), (ax2), (ax3)) = plt.subplots(1,3)
ax1.plot(scheme, comm_cost, color='blue', marker="^")
ax1.set_title('Communication Cost (in bits)', fontsize=12, fontweight='bold', **csfont)
ax1.set_xlabel('', fontsize=12, fontweight='bold', **hfont)
ax1.set_ylabel('Communication Cost', fontsize=12, fontweight='bold', **hfont)
#ax1.set_xticks(scheme, rotation=60)
#ax1.set_margins(x=0)
ax1.grid(True)
#ax1.tight_layout()

ax2.plot(scheme, comp_cost, color='red', marker="o", linestyle='--')
ax2.set_title('Computation Cost (in  ms)', fontsize=12, fontweight='bold', **csfont)
ax2.set_xlabel('', fontsize=12, fontweight='bold', **hfont)
ax2.set_ylabel('Computation Cost', fontsize=12, fontweight='bold', **hfont)
#ax2.xticks(scheme, rotation=60)
#ax2.margins(x=0)
ax2.grid(True)

ax3.plot(scheme, sec_level, color='green', marker="P", linestyle='--')
ax3.set_title('Security Level', fontsize=12, fontweight='bold', **csfont)
ax3.set_xlabel('', fontsize=12, fontweight='bold', **hfont)
ax3.set_ylabel('Security Level', fontsize=12, fontweight='bold', **hfont)
#ax2.xticks(scheme, rotation=60)
#ax2.margins(x=0)
ax3.grid(True)
#ax2.tight_layout()
#plt.savefig("comm_plot.pdf")  
#files.download("comm_plot.pdf") 
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=1.9, 
                    hspace=1.4)

plt.setp(ax1.get_xticklabels(), rotation=70, horizontalalignment='right')
plt.setp(ax2.get_xticklabels(), rotation=70, horizontalalignment='right')
plt.setp(ax3.get_xticklabels(), rotation=70, horizontalalignment='right')
plt.tight_layout()
plt.savefig("comm_plot.pdf")  
files.download("comm_plot.pdf") 
plt.show()
```
